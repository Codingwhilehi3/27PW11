### Question 1 

```python
# taking inputs 
n = int(input("Enter the number of books: "))
m = int(input("Enter the number of students: "))
pages = list(map(int, input ("Enter the number of pages in each book: ").split()))
```

```python
# defining function 
def book_allocation(pages, m, n):
    
    if m > n:
        return -1
    
    low = max(pages)
    high = sum(pages)
    result = -1
    
    while low <= high:
        mid = (low + high) // 2
        
        students_used = 1
        current_pages = 0
        can_allocate = True
        
        for page in pages:
            if page > mid:
                can_allocate = False
                break
                
            if current_pages + page > mid:
                students_used += 1
                current_pages = page
                
                if students_used > m:
                    can_allocate = False
                    break
            else:
                current_pages += page
        
        if can_allocate:
            result = mid
            high = mid - 1
        else:
            low = mid + 1
    
    return result
```

```python
book_allocation(pages, m, n)
```

```python
# taking inputs 
sum_of_coins = int(input("Enter the sum: "))
coins = list(map(int, input ("Enter the different coin denominations available: ").split()))
```

```python
# dynamic programming algorithm 

n = len(coins)
dp = [10000000] * (sum_of_coins + 1) # instead of infinity, a very large number is used
dp[0] = 0
# value ranges from 1 to sum_of_coins 
for c in coins:
    for value in range(1, sum_of_coins + 1):
        if (value - c >= 0):
            dp[value] = min(dp[value], 1 + dp[value - c])
        else:
            continue

if dp[sum_of_coins] == 10000000:
    print ("-1")
else:
    print (dp[sum_of_coins])
```

### Question 3

```python
# defining the linked list
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None
```

```python
# creating the linked list 
def create_linked_list(values):
    if not values:
        return None
    
    head = ListNode(values[0])
    current = head
    
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    
    return head
```

```python
# adds a cycle in the list 
def create_cycle(head, cycle_pos):
    if cycle_pos == -1 or not head:
        return head
    
    current = head
    cycle_node = None
    tail = None
    pos = 0
    
    while current:
        if pos == cycle_pos:
            cycle_node = current
        if not current.next:
            tail = current
        current = current.next
        pos += 1
    
    if tail and cycle_node:
        tail.next = cycle_node
    
    return head
```

```python
''' 
Floyd algorithm method - 

Time complexity - O(n)
Space complexity - O(1)

Another method is to allot a "visited = 0" variable to each node and change "visited = 1" when the pointer reaches it
Exit the loop when a node with already "visited = 1" is spotted. The time and space complexity of this algorithm is also O(n) and O(1) respectively.
But this algorithm is not preferred since this involves editing the orignal data. 
Hence, Floyd's cycle detection algorithm is used. 

'''

def has_loop(head):
    if not head or not head.next:
        return False
    
    slow = head  
    fast = head  
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:  
            return True
    
    return False
```

```python
# testing with inputs 
if __name__ == "__main__":
    
    head1 = create_linked_list([1, 2, 3, 4, 5])
    print(f"List [1,2,3,4,5] : {has_loop(head1)}")
    
    head2 = create_linked_list([3, 2, 0, -4])
    head2 = create_cycle(head2, 1)  # creates cycle at position 1
    print(f"List [3,2,0,-4] : {has_loop(head2)}")
    
    head3 = ListNode(1)
    head3.next = head3  
    print(f"Single node with self loop: {has_loop(head3)}")
    
    print(f"Empty list has cycle: {has_loop(None)}")
```
