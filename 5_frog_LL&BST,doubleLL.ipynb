### Question 1

```python
n = int(input("Enter the value of n: "))
```

```python
MOD = 1000000007
```

```python
def no_of_ways(n):
    # Base cases
    if n == 0:
        return 1
    elif n == 1:
        return 1
    elif n == 2:
        return 2

    dp = [0] * (n + 1)
    dp[0], dp[1], dp[2] = 1, 1, 2

    for i in range(3, n + 1):
        dp[i] = (dp[i-1] + dp[i-2] + dp[i-3]) % MOD

    return dp[n]
```

```python
print ("Number of ways to reach the top " + str(no_of_ways(n)))
```

```python
# Definition for a Node in the doubly linked list
class Node:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None
        self.child = None
```

```python
def flatten(head):
    if not head:
        return None

    stack = []
    curr = head

    while curr:
        # if there is a child
        if curr.child:
            # if there is a next node, push it to stack for later
            if curr.next:
                stack.append(curr.next)
            
            # connect child as the next node
            curr.next = curr.child
            curr.child.prev = curr
            curr.child = None  # remove child pointer

        # if we reach the end and there's something in stack
        if not curr.next and stack:
            curr.next = stack.pop()
            curr.next.prev = curr

        curr = curr.next

    return head
```

```python
def print_list(head):
    """Prints the flattened doubly linked list"""
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.next
    print()
```

```python
def build_sample_list():
    # Main list
    head = Node(1)
    n2 = Node(2)
    n3 = Node(3)
    n4 = Node(4)
    head.next = n2; n2.prev = head
    n2.next = n3; n3.prev = n2
    n3.next = n4; n4.prev = n3

    # First child
    n5 = Node(5)
    n6 = Node(6)
    n2.child = n5
    n5.next = n6; n6.prev = n5

    # Second child
    n7 = Node(7)
    n8 = Node(8)
    n6.child = n7
    n7.next = n8; n8.prev = n7

    return head
```

```python
if __name__ == "__main__":
    head = build_sample_list()
    print("Flattened list:")
    head = flatten(head)
    print_list(head)
```
