### Question 1

```python
# taking inputs 
n = int(input("Enter number of gas stations: "))
gas = list(map(int, input ("Enter the amount of gas at each station starting from station 0, seperated by spaces: ").split()))
cost = list(map(int, input ("Enter the amount of gas required to go to the station starting from station 0, seperated by spaces: ").split()))
```

```python
def firstQuestion(gas, cost, n):
    total = 0 
    tank = 0
    start = 0
    for i in range(0, n):
        diff = gas[i] - cost[i]
        total += diff       
        tank += diff       

        if tank < 0:
            start = i + 1  
            tank = 0
    return start if total >= 0 else -1
```

```python
firstQuestion(gas, cost, n)
```

### Question 2

```python
# taking inputs
brackets = str(input("Enter the brackets: "))
```

```python
def secondQuestion(brackets):
    stack = []
    match = {')':'(', ']':'[', '}':'{'} 
    
    for ch in brackets:
        if ch in "([{":
            stack.append(ch)
        elif ch in ")]}":
            if not stack or stack[-1] != match[ch]:
                return "Not Balanced"
            stack.pop()
    
    return "Balanced" if not stack else "Not Balanced"
```

```python
secondQuestion(brackets)
```

### Question 3

```python
# taking inputs 
first_text = input("Enter first sequence: ")
second_text = input("Enter second sequence: ")
```

```python
def thirdQuestion (first_text, second_text):
    m, n = len(first_text), len(second_text)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if first_text[i-1] == second_text[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    return dp[m][n]
```

```python
thirdQuestion (first_text, second_text)
```
