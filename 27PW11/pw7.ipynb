### Question 1

```python
class Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def build_sample_bst():
    # building tree with values and connections 
    root = Node(4)
    root.left = Node(1)
    root.right = Node(6)
    root.left.left = Node(0)
    root.left.right = Node(2)
    root.left.right.right = Node(3)
    root.right.left = Node(5)
    root.right.right = Node(7)
    root.right.right.right = Node(8)
    return root
```

```python
def BST_to_GT(root):
    node_sum = 0;
    if not root:
        return None

    def reverse_inorder(node):
            nonlocal node_sum
            if not node:
                return
            reverse_inorder(node.right) # recursion to the right
            node_sum += node.val # adding node value to node_sum 
            node.val = node_sum # updating node value
            reverse_inorder(node.left) # recursion to the right

    reverse_inorder(root)
    return root
```

```python
def print_level_order(root):
    if not root:
        return []

    from collections import deque
    queue = deque([root])
    result = []

    while queue:
        node = queue.popleft()
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)

    # Trim trailing None values
    while result and result[-1] is None:
        result.pop()

    print(result)
    return result
```

```python
if __name__ == "__main__":
    root = build_sample_bst()
    print("Greater Tree (level order):")
    root = BST_to_GT(root)
    print_level_order(root)
```

### Question 2

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
def maxResourceAllocation(root):
    def dfs(node):
        if not node:
            return (0, 0)  # (include, exclude)

        left_include, left_exclude = dfs(node.left)
        right_include, right_exclude = dfs(node.right)

        include = node.val + left_exclude + right_exclude
        exclude = max(left_include, left_exclude) + max(right_include, right_exclude)

        return (include, exclude)

    include, exclude = dfs(root)
    return max(include, exclude)
```

```python
if __name__ == "__main__":
    # value of nodes
    CEO = Node(10)
    Finance = Node(1)
    Sales = Node(2)
    HR = Node(4)
    IT = Node(5)
    Marketing = Node(3)

    # edges 
    CEO.left = Finance
    CEO.right = Sales
    Finance.left = HR
    Finance.right = IT
    Sales.right = Marketing

    # calling the function
    print(maxResourceAllocation(CEO)) 
```

### Question 3

```python
from collections import defaultdict # helps make list if lists without initial initialization 
```

```python
n = int(input("Enter the number of employees: "))

tree = defaultdict(list)
has_manager = [False] * (n + 1)

print("Enter the managerial relationships (manager employee):")
for _ in range(n - 1): # since if there are 'n' nodes, there can be at max 'n-1' edges 
    manager, employee = map(int, input().split())
    tree[manager].append(employee)
    has_manager[employee] = True

emp1, emp2 = map(int, input("Enter the two employees to find their closest common manager: ").split())
```

```python
tree # children of each manager
```

```python
has_manager # denotes whether a node as atleast one child or not
```

```python
root = None
for i in range(1, n + 1):
    if not has_manager[i]: # the first node with no parent is made the root
        root = i
        break
```

```python
def findLCA(root, emp1, emp2, tree):
    if root is None:
        return None
    
    # If current node matches one of the employees, return it
    if root == emp1 or root == emp2:
        return root
    
    found = [] # children of root appended here
    
    # DFS for all children
    for child in tree[root]:
        res = findLCA(child, emp1, emp2, tree) # recursion
        if res is not None:
            found.append(res)
    
    # If two or more children return a match, this node is LCA
    if len(found) >= 2:
        return root
    
    # Otherwise return the found node, or None
    return found[0] if found else None
```

```python
lca = findLCA(root, emp1, emp2, tree) # calling the function
print(f"Closest common manager for {emp1} and {emp2}: {lca}")
```
