### Question 1

```python
##### taking inputs
n = int(input("Enter the size of the matrix: "))
matrix = []
print("Enter the elements of the matrix:")
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
```

```python
# defining function 
def fold_matrix(matrix, n):
    if n % 2 == 0: # even size
        rows = n // 2
        for i in range(rows):
            for j in range(n):
                matrix[i][j] *= matrix[n-1-i][j]
        matrix = matrix[:rows]
    else: # odd size
        mid = n // 2
        for i in range(mid):
            for j in range(n):
                matrix[i][j] *= matrix[n-1-i][j]
        matrix = matrix[:mid+1]
    
    current_rows = len(matrix)
    if n % 2 == 0:
        cols = n // 2
        for i in range(current_rows):
            for j in range(cols):
                matrix[i][j] *= matrix[i][n-1-j]
        for i in range(current_rows):
            matrix[i] = matrix[i][:cols]
    else:
        mid = n // 2
        for i in range(current_rows):
            for j in range(mid):
                matrix[i][j] *= matrix[i][n-1-j]
        for i in range(current_rows):
            matrix[i] = matrix[i][:mid+1]
    
    return matrix
```

```python
result = fold_matrix(matrix, n)
```

### Question 2

```python
# taking inputs 
n = int(input("Enter number of guests to party: "))
entry = list(map(int, input ("Enter the entry time of the guests: ").split()))
exit = list(map(int, input ("Enter the exit time of the guests: ").split()))
```

```python
entry
```

```python
exit
```

```python
# defining function 
def guest_count (n, entry, exit):
    min_time = min(entry)
    max_time = max(exit)
    
    max_guests = 0
    best_time = 0
    
    for time in range(min_time, max_time + 1):
        current_guests = 0
        
        for i in range(n):
            if entry[i] <= time <= exit[i]:
                current_guests += 1
        
        if current_guests > max_guests:
            max_guests = current_guests
            best_time = time
    
    return best_time, max_guests
```

```python
max_time, max_guests = guest_count(n, entry, exit)
```

```python
print ("There are maximum " + str(max_guests) + " guests at time " + str(max_time))
```

```python
# taking inputs 
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter the elements of the array: ").split()))
```

```python
# defining function 
def peak_element(n, arr):
    if n == 1: # if only one element 
        return arr[0]
    
    if arr[0] > arr[1]: # to check if the first element is peak
        return arr[0]
      
    if arr[n-1] > arr[n-2]: # to check if last element is peak
        return arr[n-1]
    
    for i in range(1, n-1):
        if arr[i] > arr[i-1] and arr[i] > arr[i+1]:
            return arr[i]
    
    return -1 
```

```python
peak_element(n, arr)
```
