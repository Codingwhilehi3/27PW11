### Question 1

```python
# taking inputs 
n = int(input("Enter the number of days: "))
prices = list(map(int, input ("Enter the stock prices for each day: ").split()))
fee = int(input("Enter the transaction fee: "))
```

```python
def maxProfit(n , prices, fee):
    if not prices:
        return 0
        
    hold =- prices[0]
    not_hold = 0  

    for i in range(1, n):
        
        new_hold = max(hold, not_hold - prices[i]) # either keep holding or buy now
        new_not_hold = max(not_hold, hold + prices[i] - fee) # either keep holding or sell now
        
        hold = new_hold
        not_hold = new_not_hold
    
    return not_hold
```

```python
maxProfit(n, prices, fee)
```

### Question 2

```python
# taking inputs 
n = int(input("Length of rod in inches: "))
prices = list(map(int, input ("Enter the value for each piece of rod: ").split()))
```

```python
def maxValue(price, n):
    # initializing dp
    dp = [0] * (n + 1)
    
    # For each length from 1 to n
    for length in range(1, n + 1):
        # Try all possible first cuts
        for cut in range(1, length + 1):
            # cut has length 'cut', remaining has length 'length - cut'
            dp[length] = max(dp[length], price[cut - 1] + dp[length - cut])
    
    return dp[n]
```

```python
maxValue(prices, n)
```

### Question 3

```python
str1 = str(input("Enter first string: "))
str2 = str(input("Enter second string: "))
```

```python
def minCost(str1, str2):
    n = len(str1)
    str1 = list(str1)  # Convert to list for easier manipulation
    str2 = list(str2)  # Convert to list for easier manipulation
    
    a_to_b = []  # Positions where str1='a' but str2='b'
    b_to_a = []  # Positions where str1='b' but str2='a'
    
    for i in range(n):
        if str1[i] != str2[i]:
            if str1[i] == 'a':
                a_to_b.append(i)
            else:
                b_to_a.append(i)
    
    total_cost = 0
    
    while a_to_b and b_to_a:
        # Find the pair with minimum swap cost
        min_cost = float('inf')
        best_i, best_j = -1, -1
        
        for i in range(len(a_to_b)):
            for j in range(len(b_to_a)):
                swap_cost = abs(a_to_b[i] - b_to_a[j])
                if swap_cost < min_cost:
                    min_cost = swap_cost
                    best_i, best_j = i, j
        
        # Compare swap vs change
        if min_cost < 2:  # Swap is cheaper than 2 changes
            total_cost += min_cost
            a_to_b.pop(best_i)
            b_to_a.pop(best_j) 
        else:
            # Change is cheaper, change one character
            total_cost += 1
            a_to_b.pop(0)
    
    # Remaining mismatches need to be changed (cost = 1 each)
    total_cost += len(a_to_b) + len(b_to_a)
    
    return total_cost
```

```python
minCost(str1, str2)
```
