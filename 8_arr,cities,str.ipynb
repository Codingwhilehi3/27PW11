### Question 1

```python
n = int(input("Enter size of array: "))
arr1 = list(map(int, input("Enter the elements of the sorted array: ").split()))              
```

```python
print("The original sequence of elements is " + str(arr1))
```

```python
left = 0
right = n - 1
arr2 = []
while (left < right):
    arr2.append(arr1[right])
    arr2.append(arr1[left])
    right -= 1
    left += 1
```

```python
print("The elements rearranged in max-min form are: " + str(arr2)) 
```

### Question 2

```python
n = int(input("Enter the number of cities: "))
isConnected = []

print("Enter the isConnected matrix row by row:")

for i in range(n):
    row = list(map(int, input().split()))
    isConnected.append(row)
```

```python
def provinces(isConnected):
    n = len(isConnected)
    visited = [False] * n

    def dfs(city):
        for neighbor in range(n):
            if isConnected[city][neighbor] == 1 and not visited[neighbor]:
                visited[neighbor] = True
                dfs(neighbor)

    no_of_provinces = 0
    for city in range(n):
        if not visited[city]:
            dfs(city)
            no_of_provinces += 1

    return no_of_provinces
```

```python
print ("The number of provinces are " + str(provinces(isConnected)))
```

```python
words = list(map(str, input ("Enter the strings: ").split()))
```

```python
words
```

```python
def subStrings(words):
    result = []
    for i in range(len(words)):
        for j in range(len(words)):
            if i != j and words[i] in words[j]:
                result.append(words[i])
                break
    return result
```

```python
subStrings(words)
```
